```python
#Function 
def is_palindrome(word):
    word = word.lower()
    word = word.replace(" ", "")
    if word == word[::-1]:
        return True
    else:
        return False
    
response = is_palindrome(input("Please enter a word or phrase: "))
if response == True:
    print("Your input is a palindrome!")
else:
    print("Your input is not a palindrome...")




        
```
