```python
#Function using string slicing
def reverse_string(word):
    return word[::-1]

#result
print(reverse_string(input('Enter a string: ')))





```

```python
#Function using for loop
def reverse_string(word):
    output = ''
    for i in range(len(word)-1, -1, -1):
        output += word[i]
    return output


#result
print(reverse_string(input('Enter a word: ')))
```
