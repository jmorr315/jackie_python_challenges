```python
FIRST_UP = ord("A")
LAST_UP = ord("Z")
FIRST_LOW = ord("a")
LAST_LOW = ord("z")

def caesar_cipher(text, shift):
    result = ""

    for char in text:
        if char.isupper():  
            # Handle uppercase letters
            char_code = ord(char)
            new = char_code + shift

            if new > LAST_UP:
                new -= (LAST_UP - FIRST_UP + 1)
            if new < FIRST_UP:
                new += (LAST_UP - FIRST_UP + 1)

            result += chr(new)

        elif char.islower():
            # Handle lowercase letters
            char_code = ord(char)
            new = char_code + shift

            if new > LAST_LOW:
                new -= (LAST_LOW - FIRST_LOW + 1)
            if new < FIRST_LOW:
                new += (LAST_LOW - FIRST_LOW + 1)

            result += chr(new)

        else:
            # Non-letters get added exactly as they are
            result += char

    print(result)


user_message = input("Message to Encrypt: ")
user_shift_key = int(input("Shift Key (number): "))
caesar_cipher(user_message, user_shift_key)
```
